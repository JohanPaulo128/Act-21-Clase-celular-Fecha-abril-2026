```python
class Celular:
    # Constructor con atributos
    def __init__(self, marca, modelo, precio):
        self.marca = marca # Aseguramos que 'marca' esté inicializada
        self.modelo = modelo
        self.precio = precio
        self.estado = 'Apagado' # Atributo por defecto como string literal
        self.encendido = False # Añadimos un estado booleano para mayor claridad

    # Método para mostrar información
    def mostrar_info(self):
        return f"{self.marca} {self.modelo} - ${self.precio}"

    # Método de comportamiento
    def encender(self):
        self.estado = 'Encendido'
        self.encendido = True
        return "Celular encendido"

# Crear una instancia (objeto)
mi_celular1 = Celular("Samsung", "Galaxy S23", 800)
mi_celular2 = Celular("Xiaomi", "Redmi Note 12", 400)

# LLamando a las funciones para cada objeto celular 1 y celular 2
print("Celular 1")
print(mi_celular1.encender())
print(mi_celular1.mostrar_info())

print("Celular 2")
print(mi_celular2.encender())
print(mi_celular2.mostrar_info())

print("Programa terminado")
print("Creador de software por Johan Paulo Nl = 04")
```
